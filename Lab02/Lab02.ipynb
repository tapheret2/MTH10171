```python
# # 3. Validity
# # 3.5 Standardize misspelled data.
# # Age_Group
# df_copy.loc[df_copy['Age_Group'] == '45', 'Age_Group'] = '45-54'
# df_copy.loc[df_copy['Age_Group'] == '56', 'Age_Group'] = '>=55'
# df_copy.loc[df_copy['Age_Group'] == '22', 'Age_Group'] = '18-24'
```

```python
# Tạo danh sách giá trị hợp lệ
valid_age_groups = ['18-24', '25-34', '35-44', '45-54', '>=55']

# Chuyển cột Age_Group thành dạng chuỗi (để tránh lỗi so sánh dtype)
df_copy['Age_Group'] = df_copy['Age_Group'].astype(str)

# Chuyển các giá trị có thể thành số, giữ nguyên nếu không chuyển được
df_copy['Age_Group_Numeric'] = pd.to_numeric(df_copy['Age_Group'], errors='coerce')

# Điều kiện nhóm tuổi
conditions = [
    (df_copy['Age_Group_Numeric'] >= 18) & (df_copy['Age_Group_Numeric'] <= 24),
    (df_copy['Age_Group_Numeric'] >= 25) & (df_copy['Age_Group_Numeric'] <= 34),
    (df_copy['Age_Group_Numeric'] >= 35) & (df_copy['Age_Group_Numeric'] <= 44),
    (df_copy['Age_Group_Numeric'] >= 45) & (df_copy['Age_Group_Numeric'] <= 54),
    (df_copy['Age_Group_Numeric'] >= 55)
]

# Nhóm tuổi tương ứng
age_labels = ['18-24', '25-34', '35-44', '45-54', '>=55']

# Chỉ thay thế nếu giá trị không thuộc danh sách hợp lệ
df_copy['Age_Group'] = np.where(
    df_copy['Age_Group'].isin(valid_age_groups),  # Nếu đã hợp lệ
    df_copy['Age_Group'],  # Giữ nguyên
    np.select(conditions, age_labels, default="Unknown")  # Thay thế nếu sai
)

# Xóa cột trung gian
df_copy.drop(columns=['Age_Group_Numeric'], inplace=True)

# Kiểm tra kết quả
print(df_copy['Age_Group'].value_counts())
```
