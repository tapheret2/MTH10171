# **Exploratory Data Analysis (EDA)**

## 1. **Theory**

**What is exploratory data analysis (EDA) 🤔❓️**

✅️ **Exploratory data analysis - EDA** is used by data scientists to analyze and explore datasets and summarize the main characteristics of them. Mostly it is done by using data visualization methods.

✅️ **EDA** makes it easier for data scientists to discover patterns, spot anomalies, test a hypothesis, or check assumptions.

✅️ **EDA** is primarily used to provide a better understanding of dataset's variables and their relationships.

✅️ **EDA** can also help determine whether the statistical techniques you are considering are appropriate for data analysis.

✅️ Developed by the American mathematician John Tukey in the 1970s, EDA techniques continue to be a widely used method in the data exploration process today.

**Why is EDA so important in data science 🤔❓️**

✅️ The main purpose of EDA is to help you look at the data before making any assumptions. In addition to better understanding the patterns in the data or detecting unusual events, it also helps you find interesting relationships between variables.

✅️ Data scientists can use exploratory analysis to ensure that the results they produce are valid and relevant to desired business outcomes and goals.

✅️ EDA also helps stakeholders by verifying that they are asking the right questions.

✅️ EDA can help to answer questions about standard deviations, categorical variables, and confidence intervals.

✅️ After the exploratory analysis is completed and the predictions are determined, its features can be used for more complex data analysis or modeling, including machine learning.

**EDA techniques**


1.   Understanding the Structure of the Data:
- Data types: Check the data types of each column (e.g., numerical, categorical).
2.   Statistical Summary:
- Descriptive statistics: Get a quick summary of numerical columns (mean, median, standard deviation, min, max).
3.   Missing Data Analysis
4.   Visualizing Distributions
- Histograms: Plot histograms for numerical features to see their distribution.
- Boxplots: Boxplots help detect outliers in the dataset.
5.   Correlation Analysis
- Correlation matrix: Calculate the correlation between numerical features to understand relationships.
- Scatter plots: Visualize relationships between pairs of numerical variables.
6.   Categorical Data Analysis
7.   Numerical Data Analysis
8.   Outlier Detection
9.   Class Distribution (for Classification Problems)
10.   Multivariate Analysis
11.   Time Series Analysis (If applicable)
12. Feature Engineering and Transformation
- Scaling/Normalization: Scale numerical features to have a consistent range (useful for machine learning models).
13. Dimensionality Reduction
- PCA (Principal Component Analysis): Reduces the number of features in the dataset while retaining most of the variance.

**Skew in numerical features**

Skew refers to how asymmetrical a distribution is.

In the context of numerical features, skewness tells us whether the data is balanced around the mean or leans more to one side.



| **Skew Value**    | **Interpretation**                              |
|-------------------|-------------------------------------------------|
| ≈ 0               | Data is approximately symmetric (normal)       |
| > 0 (positive)    | Data is right-skewed (long tail to the right)  |
| < 0 (negative)    | Data is left-skewed (long tail to the left)    |

**1. Phân phối đối xứng (Skew ≈ 0)**
- Giải thích:  Khi skewness gần bằng 0, dữ liệu gần như có phân phối đối xứng (hay chuẩn), tức là phần bên trái và bên phải của trung bình (mean) có sự phân bố tương đối cân bằng.

- Ví dụ: Dữ liệu có skew ≈ 0 thường xuất hiện trong phân phối chuẩn, ví dụ như chiều cao của một nhóm người.

- Biểu hiện:
 + Dữ liệu phân bố đều ở hai phía của trung bình.
 + Mean = Median = Mode


**2. Phân phối lệch phải (Skew > 0) — Right-skewed**
- Giải thích: Khi skewness lớn hơn 0 (dương), dữ liệu có đuôi dài ở phía bên phải (right tail), tức là phần lớn dữ liệu tập trung ở bên trái của trung bình, nhưng có một số ít giá trị cực cao kéo dài phân phối ra phía phải.
- Ví dụ: Thu nhập của một nhóm người, vì có một số ít người có thu nhập cực kỳ cao (ví dụ: tỷ phú), nhưng đa phần người có thu nhập thấp hoặc trung bình.
- Biểu hiện:

 +  Dữ liệu tập trung ở phía trái, với đuôi dài ở phía phải.

 + Mean > Median > Mode.

**3. Phân phối lệch trái (Skew < 0) — Left-skewed**

 - Giải thích: Khi skewness nhỏ hơn 0 (âm), dữ liệu có đuôi dài ở phía bên trái (left tail), tức là phần lớn dữ liệu tập trung ở phía phải của trung bình, nhưng có một số ít giá trị cực thấp kéo dài phân phối ra phía trái.

- Ví dụ: Tuổi nghỉ hưu của một nhóm người, vì phần lớn người sẽ nghỉ hưu ở độ tuổi cao hơn (60-65), nhưng một số ít người nghỉ hưu sớm ở độ tuổi 30-40.

- Biểu hiện:
 + Dữ liệu tập trung ở phía phải, với đuôi dài ở phía trái.
 + Mean < Median < Mode.

**Outlier Detection**
 - Z-Score: Identify outliers using Z-scores. A Z-score above 3 or below -3 may indicate an outlier.
 - IQR (Interquartile Range): Use IQR to find and remove outliers.

```python
# Boxplot for CREDIT_SCORE
sns.boxplot(x=numerical_data['CREDIT_SCORE'])

plt.title('Boxplot of CREDIT_SCORE')
plt.xlabel('CREDIT_SCORE')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(numerical_data['CREDIT_SCORE'].dropna())
plt.title('Boxplot of CREDIT_SCORE')
plt.ylabel('CREDIT_SCORE')
plt.show()
```

```python
# Z-Score: Identify outliers using Z-scores. A Z-score above 3 or below -3 may indicate an outlier.
from scipy.stats import zscore
numerical_data_copy = numerical_data.copy()

# Handling NaN in CREDIT_SCORE
numerical_data_copy = numerical_data_copy[numerical_data_copy['CREDIT_SCORE'].notna()]

# Compute z-score
numerical_data_copy['Z_SCORE'] = zscore(numerical_data_copy['CREDIT_SCORE'])
```

```python
numerical_data_copy
```

```python
Q1 = numerical_data_copy['CREDIT_SCORE'].quantile(0.25)
Q3 = numerical_data_copy['CREDIT_SCORE'].quantile(0.75)
IQR = Q3 - Q1
numerical_data_after_removing_outlier = numerical_data_copy[(numerical_data_copy['CREDIT_SCORE'] >= (Q1 - 1.5 * IQR)) & (numerical_data_copy['CREDIT_SCORE'] <= (Q3 + 1.5 * IQR))]
```

```python
numerical_data_after_removing_outlier
```

```python
# Boxplot for CREDIT_SCORE after removing outlier
sns.boxplot(x=numerical_data_after_removing_outlier['CREDIT_SCORE'])

plt.title('Boxplot of CREDIT_SCORE')
plt.xlabel('CREDIT_SCORE')
plt.show()
```

```python
# Correlation matrix of whole dataset
# Generating correlation coefficient between each features
from pandas import set_option

# Set display precision
# Set the precision for float display
set_option('display.float_format', '{:,.3f}'.format)

# Ensure we're working with only numeric columns

# Calculate correlation matrix
correlation = numerical_data.corr(method='pearson')

# Display the correlation matrix
correlation
```

```python
# Plot correlation matrix in heatmap
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(numerical_data.corr(), ax=ax, annot=True)
plt.show()
```

```python
# Correlation plot between numerical features and target
fig, ax = plt.subplots(3,1, figsize=(12, 12))
## Correlation coefficient using different methods
corr1 = numerical_data.corr('pearson')[['OUTCOME']].sort_values(by='OUTCOME', ascending=False)
corr2 = numerical_data.corr('spearman')[['OUTCOME']].sort_values(by='OUTCOME', ascending=False)
corr3 = numerical_data.corr('kendall')[['OUTCOME']].sort_values(by='OUTCOME', ascending=False)

#setting titles for each plot
ax[0].set_title('Pearson method')
ax[1].set_title('spearman method')
ax[2].set_title('Kendall method')
## Generating heatmaps of each methods
sns.heatmap(corr1, ax=ax[0], annot=True)
sns.heatmap(corr2, ax=ax[1], annot=True)
sns.heatmap(corr3, ax=ax[2], annot=True)

plt.show()
```

## 2.4. Check imbalance in the data

```python
#check Imbalance in data

#group instances based on the classes in OUTCOME variable
class_counts=df_train.groupby("OUTCOME").size()

columns=['outcome','count','percentage']
outcome=[0,1]
count=list()
percentage=list()

#Calculate the percentage of each value of the OUTCOME variable from total
for val in range(2):
    count.append(class_counts[val])
    percent=(class_counts[val]/105000)*100
    percentage.append(percent)

# Convert the calulated values into a dataframe
imbalance_df=pd.DataFrame(list(zip(outcome,count,percentage)),columns=columns)
imbalance_df
```

```python
sns.barplot(data=imbalance_df,x=imbalance_df['outcome'],y=imbalance_df['percentage'])
plt.show()
```

**Remarks:**
- Based on the above grouping, It shows that around 58% of instances consists OUTCOME value '0' and 42% of instances consists OUTCOME values '1'
